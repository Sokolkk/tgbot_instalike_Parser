#Структура HTML-страницы 
**Язык гипертекстовой разметки **(HTML) - это стандартный язык разметки для документов, предназначенных для отображения в веб-браузере. 

HTML описывает структуру веб-страницы и может использоваться с каскадными таблицами стилей (CSS) и языком сценариев, например JavaScript, для создания интерактивных веб-сайтов. HTML состоит из ряда элементов, которые «сообщают» браузеру, как отображать контент. Наконец, элементы представлены тегами. 

Вот несколько тегов: 

<! DOCTYPE html> Объявление определяет этот документ как HTML5. 

< html>  - это корневой элемент HTML-страницы.

< div> тег определяет подразделение или раздел в документе HTML. Обычно это контейнер для других элементов. 

< head> Элемент содержит метаинформацию о документе. 

< title> Элемент определяет заголовок документа. 

< body> элемент содержит видимое содержимое страницы. 

< h1> элемент определяет большой заголовок. 

< p> элемент определяет абзац. 

< a> элемент определяет гиперссылку. 

HTML - теги обычно приходят парами , как < p> и </ p>. Первый тег в паре - это открывающий тег, второй тег - закрывающий. Конечный тег записывается как начальный, но перед именем тега вставляется косая черта.

Мы будем использовать библиотеки requests и BeautifulSoup для доступа и очистки содержимого нужного нам сайта.

BeautifulSoup это библиотека Python для извлечения данных из файлов HTML и XML. Он предоставляет методы для навигации по древовидной структуре документа, о которой мы говорили ранее, и для очистки ее содержимого.

**То есть шаги будут следующие:**

1.   с помощью requests получим доступ к странице
2.   BeautifulSoup спарсит HTML документ
3.   Chrome DevTools обнаружит что извлечь
4.   BeautifulSoup достанет теги
5.   после сможем извлечь контент со страницы



```python
# делаем импорты
import numpy as np #библиотека для работы с данными
import requests #запросы
from bs4 import BeautifulSoup #парсер
import pandas as pd #библиотека для работы с данными
import matplotlib.pyplot as plt #работа с визуализацией
%matplotlib inline 
# matplotlib inline заставит ваши выходные данные графика появиться и сохраниться в файле ноутбука.
```

```python
# указываем адрес страницы для извлечения
url = 'https://proglib.io/'

# Используем requests для обращения к URL
response = requests.get(url)

# BeautifulSoup спарсит HTML документ
soup = BeautifulSoup(response.text, 'html.parser')

# Заголовок полученной страницы
soup.title
```

```python
# Мы также можем получить это без HTML-тегов <title> </title> используя метод .string
soup.title.string
```

```python
#мы можем получить, например, заголовок видео найдя его в элементах и прописав поиск супу через метод find, где сначала пишется тег, используемый элементом, после название класса и получаем все его содержимое
card = soup.find('div', {'class': 'preview-card__content'})
print(card)#если суп находит на странице то выдает результат, если нет, то NONE
```

image.png

```python
#получаем все заголовки статей h2 циклом
for article in soup.find_all('div', {'class': 'preview-card__content'}): #указываем тот же класс, что и ранее
    if article.h2:  #если оно есть, то вывод делаем, если нет, то вывода не будет
        print(article.h2.string)
 
```

```python
# находим и выводим все ссылки на странице
links = [link.get('href') for link in soup.find_all('a')]
print(links)
# Добавьте домашнюю страницу и сохраните уникальные ссылки
fixed_links = set([''.join([url, link]) for link in links if link])
print(fixed_links)
#на выходе получаем 2 строчки со ссылками
```
