DevTools –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å –≤–∞–º –ø—Ä–æ—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å –∏ —Ä–µ–¥–∞–∫—Ç–∏—Ä–æ–≤–∞—Ç—å –≤–µ–±-—Å—Ç—Ä–∞–Ω–∏—Ü—ã. –ú—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç Chrome, —á—Ç–æ–±—ã –ø—Ä–æ–≤–µ—Ä–∏—Ç—å HTML-—Å—Ç—Ä–∞–Ω–∏—Ü—É –∏ –Ω–∞–π—Ç–∏, –∫–∞–∫–∏–µ —ç–ª–µ–º–µ–Ω—Ç—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –¥–∞–Ω–Ω—ã–º, –∫–æ—Ç–æ—Ä—ã–µ –º—ã, –≤–æ–∑–º–æ–∂–Ω–æ, –∑–∞—Ö–æ—Ç–∏–º –æ—á–∏—Å—Ç–∏—Ç—å.

–ß—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –Ω–µ–∫–æ—Ç–æ—Ä—ã–π –æ–ø—ã—Ç —Ä–∞–±–æ—Ç—ã —Å–æ —Å—Ç—Ä—É–∫—Ç—É—Ä–æ–π HTML-—Å—Ç—Ä–∞–Ω–∏—Ü—ã –∏ Chrome DevTools, –º—ã –±—É–¥–µ–º –∏—Å–∫–∞—Ç—å –∏ –Ω–∞—Ö–æ–¥–∏—Ç—å —ç–ª–µ–º–µ–Ω—Ç—ã –≤ proglib.io

–í—ã –º–æ–∂–µ—Ç–µ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –∫–æ–¥ –∏ —Å—Ç—Ä—É–∫—Ç—É—Ä—É –ª—é–±–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã –Ω–∞–∂–∞–≤ f12 –Ω–∞ –∫–ª–∞–≤–∏–∞—Ç—É—Ä–µ, –ø–æ–ª—É—á–∏–≤ –¥–æ—Å—Ç—É–ø –∫ –ø–∞–Ω–µ–ª–∏ —ç–ª–µ–º–µ–Ω—Ç–æ–≤.

–ú—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ requests –∏ BeautifulSoup –¥–ª—è –¥–æ—Å—Ç—É–ø–∞ –∏ –æ—á–∏—Å—Ç–∫–∏ —Å–æ–¥–µ—Ä–∂–∏–º–æ–≥–æ –Ω—É–∂–Ω–æ–≥–æ –Ω–∞–º —Å–∞–π—Ç–∞.

BeautifulSoup —ç—Ç–æ –±–∏–±–ª–∏–æ—Ç–µ–∫–∞ Python –¥–ª—è –∏–∑–≤–ª–µ—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Ñ–∞–π–ª–æ–≤ HTML –∏ XML. –û–Ω –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª—è–µ—Ç –º–µ—Ç–æ–¥—ã –¥–ª—è –Ω–∞–≤–∏–≥–∞—Ü–∏–∏ –ø–æ –¥—Ä–µ–≤–æ–≤–∏–¥–Ω–æ–π —Å—Ç—Ä—É–∫—Ç—É—Ä–µ –¥–æ–∫—É–º–µ–Ω—Ç–∞, –æ –∫–æ—Ç–æ—Ä–æ–π –º—ã –≥–æ–≤–æ—Ä–∏–ª–∏ —Ä–∞–Ω–µ–µ, –∏ –¥–ª—è –æ—á–∏—Å—Ç–∫–∏ –µ–µ —Å–æ–¥–µ—Ä–∂–∏–º–æ–≥–æ.

**–¢–æ –µ—Å—Ç—å —à–∞–≥–∏ –±—É–¥—É—Ç —Å–ª–µ–¥—É—é—â–∏–µ:**

1.   —Å –ø–æ–º–æ—â—å—é requests –ø–æ–ª—É—á–∏–º –¥–æ—Å—Ç—É–ø –∫ —Å—Ç—Ä–∞–Ω–∏—Ü–µ
2.   BeautifulSoup —Å–ø–∞—Ä—Å–∏—Ç HTML –¥–æ–∫—É–º–µ–Ω—Ç
3.   Chrome DevTools –æ–±–Ω–∞—Ä—É–∂–∏—Ç —á—Ç–æ –∏–∑–≤–ª–µ—á—å
4.   BeautifulSoup –¥–æ—Å—Ç–∞–Ω–µ—Ç —Ç–µ–≥–∏
5.   –ø–æ—Å–ª–µ —Å–º–æ–∂–µ–º –∏–∑–≤–ª–µ—á—å –∫–æ–Ω—Ç–µ–Ω—Ç —Å–æ —Å—Ç—Ä–∞–Ω–∏—Ü—ã



```python
# –¥–µ–ª–∞–µ–º –∏–º–ø–æ—Ä—Ç—ã
import numpy as np #–±–∏–±–ª–∏–æ—Ç–µ–∫–∞ –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –¥–∞–Ω–Ω—ã–º–∏
import requests #–∑–∞–ø—Ä–æ—Å—ã
from bs4 import BeautifulSoup #–ø–∞—Ä—Å–µ—Ä
import pandas as pd #–±–∏–±–ª–∏–æ—Ç–µ–∫–∞ –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –¥–∞–Ω–Ω—ã–º–∏
import matplotlib.pyplot as plt #—Ä–∞–±–æ—Ç–∞ —Å –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–µ–π
%matplotlib inline 
# matplotlib inline –∑–∞—Å—Ç–∞–≤–∏—Ç –≤–∞—à–∏ –≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∞ –ø–æ—è–≤–∏—Ç—å—Å—è –∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å—Å—è –≤ —Ñ–∞–π–ª–µ –Ω–æ—É—Ç–±—É–∫–∞.
```

```python
#¬†—É–∫–∞–∑—ã–≤–∞–µ–º –∞–¥—Ä–µ—Å —Å—Ç—Ä–∞–Ω–∏—Ü—ã –¥–ª—è –∏–∑–≤–ª–µ—á–µ–Ω–∏—è
url = 'https://proglib.io/'

# –ò—Å–ø–æ–ª—å–∑—É–µ–º requests –¥–ª—è –æ–±—Ä–∞—â–µ–Ω–∏—è –∫ URL
response = requests.get(url)

# BeautifulSoup —Å–ø–∞—Ä—Å–∏—Ç HTML –¥–æ–∫—É–º–µ–Ω—Ç
soup = BeautifulSoup(response.text, 'html.parser')

#¬†–ó–∞–≥–æ–ª–æ–≤–æ–∫ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã
soup.title
```

```python
# –ú—ã —Ç–∞–∫–∂–µ –º–æ–∂–µ–º –ø–æ–ª—É—á–∏—Ç—å —ç—Ç–æ –±–µ–∑ HTML-—Ç–µ–≥–æ–≤ <title> </title> –∏—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥ .string
soup.title.string
```

```python
#–º—ã –º–æ–∂–µ–º –ø–æ–ª—É—á–∏—Ç—å, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∑–∞–≥–æ–ª–æ–≤–æ–∫ –≤–∏–¥–µ–æ –Ω–∞–π–¥—è –µ–≥–æ –≤ —ç–ª–µ–º–µ–Ω—Ç–∞—Ö –∏ –ø—Ä–æ–ø–∏—Å–∞–≤ –ø–æ–∏—Å–∫ —Å—É–ø—É —á–µ—Ä–µ–∑ –º–µ—Ç–æ–¥ find, –≥–¥–µ —Å–Ω–∞—á–∞–ª–∞ –ø–∏—à–µ—Ç—Å—è —Ç–µ–≥, –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã–π —ç–ª–µ–º–µ–Ω—Ç–æ–º, –ø–æ—Å–ª–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–ª–∞—Å—Å–∞ –∏ –ø–æ–ª—É—á–∞–µ–º –≤—Å–µ –µ–≥–æ —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ
card = soup.find('div', {'class': 'preview-card__content'})
print(card)#–µ—Å–ª–∏ —Å—É–ø –Ω–∞—Ö–æ–¥–∏—Ç –Ω–∞ —Å—Ç—Ä–∞–Ω–∏—Ü–µ —Ç–æ –≤—ã–¥–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –µ—Å–ª–∏ –Ω–µ—Ç, —Ç–æ NONE
```

```python
#–ø–æ–ª—É—á–∞–µ–º –≤—Å–µ –∑–∞–≥–æ–ª–æ–≤–∫–∏ —Å—Ç–∞—Ç–µ–π h2 —Ü–∏–∫–ª–æ–º
for article in soup.find_all('div', {'class': 'preview-card__content'}): #—É–∫–∞–∑—ã–≤–∞–µ–º —Ç–æ—Ç –∂–µ –∫–ª–∞—Å—Å, —á—Ç–æ –∏ —Ä–∞–Ω–µ–µ
    if article.h2:  #–µ—Å–ª–∏ –æ–Ω–æ –µ—Å—Ç—å, —Ç–æ –≤—ã–≤–æ–¥ –¥–µ–ª–∞–µ–º, –µ—Å–ª–∏ –Ω–µ—Ç, —Ç–æ –≤—ã–≤–æ–¥–∞ –Ω–µ –±—É–¥–µ—Ç
        print(article.h2.string)
 
```

```python
# –Ω–∞—Ö–æ–¥–∏–º –∏ –≤—ã–≤–æ–¥–∏–º –≤—Å–µ —Å—Å—ã–ª–∫–∏ –Ω–∞ —Å—Ç—Ä–∞–Ω–∏—Ü–µ
links = [link.get('href') for link in soup.find_all('a')]
print(links)
#¬†–î–æ–±–∞–≤—å—Ç–µ –¥–æ–º–∞—à–Ω—é—é —Å—Ç—Ä–∞–Ω–∏—Ü—É –∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å—Å—ã–ª–∫–∏
fixed_links = set([''.join([url, link]) for link in links if link])
print(fixed_links)
#–Ω–∞ –≤—ã—Ö–æ–¥–µ –ø–æ–ª—É—á–∞–µ–º 2 —Å—Ç—Ä–æ—á–∫–∏ —Å–æ —Å—Å—ã–ª–∫–∞–º–∏
```
