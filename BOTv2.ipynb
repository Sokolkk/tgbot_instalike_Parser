Если вы собираетесь использовать своего бота в группе Telegram, вам необходимо понимать, как работает конфиденциальность в Telegram. По умолчанию бот не будет получать все сообщения в групповом чате. Согласно официальной документации, он получит только:
Командные сообщения, начинающиеся с /символа косой черты (например, /help).
-Отвечает на сообщения, отправленные вашим ботом.
-Служебные сообщения (добавление или удаление людей из группы).
-Сообщения из каналов, участником которых является бот.

```python
import telebot #импортируем модуль для работы с телеграм апи
import random #импортируем модуль для работы со случайными числами
from datetime import datetime #импортируем модуль для работы со временем

#Создаем новый TeleBot экземпляр, передав токен Telegram, который мы создали ранее
bot = telebot.TeleBot("ТокенБота")
rand = random.randint(1,111) #создаем переменную для случайных чисел

now = datetime.now() # текущие дата и время

year = now.strftime("%Y") # текущий год
print("year:", year) #выводим для отображения при исполнении кода

month = now.strftime("%m") # текущий месяц
print("month:", month) #выводим для отображения при исполнении кода

day = now.strftime("%d") # текущий день
print("day:", day) #выводим для отображения при исполнении кода

time = now.strftime("%H:%M:%S")  # текущее время
print("time:", time) #выводим для отображения при исполнении кода

date_time = now.strftime("%m/%d/%Y, %H:%M:%S") #время и дата
print("date and time:",date_time)	 #выводим для отображения при исполнении кода

# вводим для начала диалога, /start
@bot.message_handler(commands=['start']) #прописываем комманду на которую будет реакция
def handle_command(message): #прописываем функцию реакции ответа на сообщение
    bot.reply_to(message, "Привет! Я ваш первый бот!") 


# Получаем время, /now
@bot.message_handler(commands=['now'])
def handle_command(message):
    bot.reply_to(message, 'Сейчас ' + str(now))#время сейчас, делаем переменную now строкой, чтобы не было ошибки сложения строки и даты


# Получаем год, /year
@bot.message_handler(commands=['year'])
def handle_command(message):
    bot.reply_to(message, 'Сейчас ' + str(year)  + ' год')#год сейчас, делаем переменную year строкой, чтобы не было ошибки сложения строки и даты

# Получаем месяц, /month
@bot.message_handler(commands=['month'])
def handle_command(message):
    bot.reply_to(message, 'Сейчас ' + str(month)  + ' месяц')#месяц сейчас, делаем переменную month строкой, чтобы не было ошибки сложения строки и даты

# Получаем день, /day
@bot.message_handler(commands=['day'])
def handle_command(message):
    bot.reply_to(message, 'Сейчас ' + str(day) + ' день')#день сейчас, делаем переменную day строкой, чтобы не было ошибки сложения строки и даты


# Получаем текущее время, /time
@bot.message_handler(commands=['time'])
def handle_command(message):
    bot.reply_to(message, 'Сейчас ' + str(time) + ' текущее время')#текущее время сейчас, делаем переменную time строкой, чтобы не было ошибки сложения строки и даты

# Получаем время и дату, /date_time
@bot.message_handler(commands=['date_time'])
def handle_command(message):
    bot.reply_to(message, 'Сейчас ' + str(date_time) + ' текущее время и дата')# текущее время и дата, делаем переменную date_time строкой, чтобы не было ошибки сложения строки и даты


# Получаем случайное число, /number
@bot.message_handler(commands=['number'])
def handle_command(message):
    bot.reply_to(message, rand)#указываем переменную со случайным числом

# все остальные сообщения  
@bot.message_handler(func=lambda message: True)
def handle_all_message(message):
	bot.reply_to(message, "Извините, я не понял, что вы имеете в виду")

bot.polling()
#polling создает новый поток, который вызывает внутреннюю функцию для автоматического получения обновлений и передачи сообщений вашим обработчикам сообщений.
#Это функция блокировки, что означает, что коды ниже не будут выполняться. Обязательно поместите его в конец файла. 
#Также не вызывайте его более одного раза, иначе произойдет ошибка. 
```
