@BotFather запросит ввести имя и имя пользователя для вашего бот, учтите, что username может быть занято, поэтому имя должно быть уникальным.

name: Отображаемое имя вашего бота.
username: Уникальный идентификатор всех ботов в Telegram. Вы не можете использовать существующие имена, которые уже есть в системе.  

После того как имя будет выбрано бот предоставит секретный токен. Будет сгенерирован новый токен для доступа к HTTP API. Обязательно держите его в безопасности, чтобы другие люди не могли использовать его для управления вашим ботом.

Если вы собираетесь использовать своего бота в группе Telegram, вам необходимо понимать, как работает конфиденциальность в Telegram. По умолчанию бот не будет получать все сообщения в групповом чате. Согласно официальной документации, он получит только:
Командные сообщения, начинающиеся с /символа косой черты (например, /help).
-Отвечает на сообщения, отправленные вашим ботом.
-Служебные сообщения (добавление или удаление людей из группы).
-Сообщения из каналов, участником которых является бот.

Чтобы взаимодействовать с ботом в чате обращаясь к нему через @ нужно:

Написать сообщение @BotFather 
/mybots
Будет получен список ваших ботов. Выбрать бота. Откроются дополнительные параметры. Нажать Bot Settings. Щелкните по кнопке Group Privacy.

Отключите настройку конфиденциальности группы, нажав на кнопку Turn off. Теперь у вашего бота будет доступ к сообщениям, отправляемым в групповых чатах. Если ваш бот уже находится в группе, новые изменения могут не отображаться и не распространяться. Если у вас возникла проблема с ним, просто удалите его из группы и снова добавьте в группу.

Бота можно будет добавлять в группы, через поиск вашего бота в списке контактов. /Start начало диалога с ним. Далее нажмите имя бота вверху. После тройные точки в верхнем правом углу. Коснитесь Add to group для выбора группы. После того как добавите бота в группу проверьте настройки группы, чтобы убедиться, что у вашего бота есть доступ к групповым сообщениям.
