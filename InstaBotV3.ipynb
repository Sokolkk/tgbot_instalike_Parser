Если Instagram обнаружит этот трюк, он заблокирует ваши действия в Instagram, такие как лайк, комментарий, подписка, отмена подписки и прямое сообщение.

##Бот позволяет установить ежедневные лимиты действий

```python
session.set_quota_supervisor(enabled=True, sleep_after=["likes", "comments_d", "follows", "unfollows", "server_calls_h"], sleepyhead=True, stochastic_flow=True, notify_me=True,
                              peak_likes=(57, 585),
                               peak_comments=(21, 182),
                                peak_follows=(48, None),
                                 peak_unfollows=(35, 402),
                                  peak_server_calls=(None, 4700))
```

**enabled**:   True для активации или False чтобы отключить 

**sleep_after**: используется для перевода InstaPy в спящий режим после достижения пика, а не для перехода к действию
Любое действие может быть включено ["likes", "comments", "follows", "unfollows", "server_calls"].

**sleepyhead**: может помочь выглядеть более человечно, чтобы просыпается немного позже в случайно выбранном временном интервале около точного времени пробуждения.

**stochastic_flow**: может обеспечить плавное генерирование пикового значения по вашим исходным значениям.

**notify_me**: Посылает уведомления (непосредственно в вашу ОС) о тех важных состояниях супервизора - сне, пробуждении и тп.

Далее устанавливаются значения для лайков, комментариев и прочих действий, где первое значение в скобках лимит в час, лимит в день.
